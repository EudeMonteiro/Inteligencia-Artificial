```python
from busca_sem_com_informacao import *
import matplotlib.pyplot as plt
from textwrap import fill

def mostrar_grafico_tempo():

   
  teste_bfs = []
  teste_dfs = []

  for i in range(100):
    teste_bfs.append(main('largura')[1])
    teste_dfs.append(main('profundidade')[1])

  bfs_tempo = np.array(teste_bfs).mean()
  dfs_tempo = np.array(teste_dfs).mean()

  nomes=['Busca em\nLargura', 'Busca em\nProfundidade']
  spc = np.arange(1, 1.5)


  plt.figure(figsize=(6, 10))
  plt.suptitle(fill("Comparação de tempo entre os algoritmos de "
                    "busca em largura e busca em profundidade", 45), y=1.11)
  
 
  ax = plt.subplot(111)

  #Remove topo e direita do gráfico  
  ax.spines['right'].set_visible(False)
  ax.spines['top'].set_visible(False)

  plt.bar(1, bfs_tempo, width=0.35)
  plt.bar(1.5, dfs_tempo, width=0.35)

  plt.rcParams.update({'font.size': 20})
  plt.ylabel("Tempo em segundos", labelpad=20)

  plt.xticks([1,1.5], nomes)
  plt.ylim(0, 0.001)

  plt.tight_layout()
  
  plt.show()

mostrar_grafico_tempo()
```
